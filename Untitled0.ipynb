```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLP resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # This line was added to download the necessary data for sentence tokenization.

```

```python
df = pd.read_csv("Amazon_Reviewssample.csv", usecols=["Score", "Summary", "Text", "Time"])
```

```python
# Convert Unix timestamp to human-readable date
df["Time"] = pd.to_datetime(df["Time"], unit="s")

print(df[["Time"]].head())  # Check if conversion is correct
```

```python
# Convert Score into Sentiment Labels
def score_to_sentiment(score):
    if score <= 2:
        return "Negative"
    elif score == 3:
        return "Neutral"
    else:
        return "Positive"

df['Sentiment'] = df['Score'].apply(score_to_sentiment)
```

```python
# Text Cleaning Function
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()  # Convert to lowercase
        text = re.sub(r'\W', ' ', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
        return text
    return ""

df['Cleaned_Text'] = df['Text'].apply(clean_text)
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    words = word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)

df['Processed_Text'] = df['Cleaned_Text'].apply(preprocess_text)
```

```python
# Save the processed dataset
df.to_csv("Processed_Amazon_Reviews(final).csv", index=False)

```

```python
# Download VADER for sentiment analysis
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Load Multi-Emotion Model
emotion_pipeline = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=3, truncation=True)
```

```python
# Function to classify multiple emotions
def detect_emotions(text):

    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return pd.Series(["None", "Low", "0", 0])

    # Get VADER Sentiment Score
    sentiment_score = sia.polarity_scores(text)['compound']

    # Get Multiple Emotion Labels (Truncated for Transformer Compatibility)
    emotions = emotion_pipeline(text[:512])[0]  # Safely truncate to 512 tokens
    emotion_labels = ", ".join([e['label'] for e in emotions])  # Store as a comma-separated string
    confidence_scores = ", ".join([str(round(e['score'], 3)) for e in emotions])  # Store confidence values
    activation_levels = ", ".join(["High" if e['score'] > 0.75 else "Medium" if e['score'] > 0.5 else "Low" for e in emotions])

    return pd.Series([emotion_labels, activation_levels, confidence_scores, sentiment_score])

# Apply Multi-Emotion Detection outside the function definition
df[['Detected_Emotions', 'Activation_Levels', 'Confidence_Scores', 'Sentiment_Score']] = df['Processed_Text'].apply(detect_emotions)
```

```python
import pandas as pd
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer

# Load dataset
df = pd.read_csv("Advanced_Emotion_Detected_Amazon_Reviews(final).csv")

# Ensure column exists
if "Processed_Text" not in df.columns:
    raise ValueError("Processed_Text column missing in dataset")

# Remove empty, NaN, or short texts
df = df[df["Processed_Text"].notna()]
df["Processed_Text"] = df["Processed_Text"].astype(str)
df = df[df["Processed_Text"].str.strip() != ""]
```

```python
# Use Sentence-Transformer Embeddings for Better Topic Representation
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")  # Efficient yet powerful model

# Configure BERTopic
bertopic_model = BERTopic(
    embedding_model=embedding_model,  # Pretrained sentence embeddings
    nr_topics="auto",  # Auto-detect the number of topics dynamically
    low_memory=True,
    calculate_probabilities=True
)

# Fit BERTopic to extract subtopics
topics, _ = bertopic_model.fit_transform(df["Processed_Text"].tolist())

# Convert numerical topic IDs into descriptive topic labels
sub_topic_labels = [
    ", ".join([word for word, _ in bertopic_model.get_topic(topic_id)])
    if topic_id != -1 else "Unknown"
    for topic_id in topics
]

# Assign extracted subtopics to DataFrame
df["Sub_Topic"] = sub_topic_labels

# Save Updated Dataset
df.to_csv("Final_Subtopics_Detected.csv", index=False)
```

```python
from transformers import pipeline

# Load the Zero-Shot Classification model
topic_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Define high-level topics
possible_topics = [
    "Product Quality", "Durability", "Defective Item", "Wrong Item Sent", "Packaging Issue",
    "Late Delivery", "Fast Delivery", "Damaged During Shipping",
    "Customer Support Response", "Refund & Returns", "Replacement Process",
    "Expensive Product", "Discounts & Deals", "Value for Money",
    "Website Usability", "Trust in Brand", "Fake or Counterfeit Product"
]

# Function to classify the main topic
def classify_main_topic(text):
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        return "Unknown"  # Handle empty text safely

    classification = topic_classifier(text, possible_topics, multi_label=True)
    topics = classification["labels"]
    scores = classification["scores"]

    # Select the most confident topic (if score > 0.5)
    if scores[0] > 0.5:
        return topics[0]
    return "General Feedback"

# Apply classification to all rows
df["Main_Topic"] = df["Processed_Text"].apply(classify_main_topic)
```

```python
from bertopic import BERTopic

# Initialize BERTopic
bertopic_model = BERTopic(nr_topics=20, low_memory=True)

# Process in batches (for large datasets)
batch_size = 10000
df["Sub_Topic"] = ""

for i in range(0, len(df), batch_size):
    batch = df.iloc[i:i+batch_size]

    # Fit BERTopic on this batch
    topics, _ = bertopic_model.fit_transform(batch["Processed_Text"].astype(str))

    # Convert topics to readable labels
    sub_topic_labels = [
        ", ".join([word for word, _ in bertopic_model.get_topic(topic_id)])
        if topic_id != -1 else "Unknown"
        for topic_id in topics
    ]

    # Assign to DataFrame
    df.loc[i:i+batch_size-1, "Sub_Topic"] = sub_topic_labels
```

```python
# Identify rows with "Unknown" subtopics
unknown_rows = df[df["Sub_Topic"] == "Unknown"]["Processed_Text"].astype(str).tolist()

if len(unknown_rows) > 0:
    print(f"Reprocessing {len(unknown_rows)} unknown subtopics...")

    # Generate embeddings for these rows
    unknown_embeddings = bertopic_model._extract_embeddings(unknown_rows, method="document")

    # Predict subtopics again
    new_topics, _ = bertopic_model.transform(unknown_rows)

    # Convert topics to readable labels
    new_topic_labels = [
        ", ".join([word for word, _ in bertopic_model.get_topic(topic_id)])
        if topic_id != -1 else "General Feedback"
        for topic_id in new_topics
    ]

    # Assign to DataFrame
    df.loc[df["Sub_Topic"] == "Unknown", "Sub_Topic"] = new_topic_labels

# Save the final results
df.to_csv("Final_Main_Subtopics.csv", index=False)
from google.colab import files
files.download("Final_Main_Subtopics.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def read_and_process_data(file_path):
    # Read CSV file
    df = pd.read_csv(file_path)

    # Split emotions, activation levels, and confidence scores
    df['Detected_Emotions'] = df['Detected_Emotions'].str.split(', ')
    df['Activation_Levels'] = df['Activation_Levels'].str.split(', ')
    df['Confidence_Scores'] = df['Confidence_Scores'].str.split(', ').apply(lambda x: list(map(float, x)))

    # Process data into structured format
    processed_data = []
    for _, row in df.iterrows():
        for emotion, activation, confidence in zip(row['Detected_Emotions'], row['Activation_Levels'], row['Confidence_Scores']):
            processed_data.append({'Emotion': emotion, 'Activation': activation, 'Confidence': confidence})

    # Convert to DataFrame
    emotion_df = pd.DataFrame(processed_data)

    # Group emotions by activation levels and calculate mean confidence score
    grouped = emotion_df.groupby(['Activation', 'Emotion'])['Confidence'].mean().unstack(fill_value=0)

    return grouped

def plot_radar_chart(data, activation_level, title):
    # Extract emotions and values
    labels = data.columns
    values = data.loc[activation_level].values

    # Radar chart setup
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # Close the radar chart
    values = np.concatenate((values, [values[0]]))
    angles += angles[:1]

    # Plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.fill(angles, values, color='blue', alpha=0.25)
    ax.plot(angles, values, color='blue', linewidth=2)

    # Labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)
    ax.set_yticklabels([])

    plt.title(title)
    plt.show()

# File path (modify if needed)
file_path = "/content/Final_Main_Subtopics.csv"

# Read and process data
grouped_data = read_and_process_data(file_path)

# Plot radar charts
plot_radar_chart(grouped_data, "High", "High Activation Emotions")
plot_radar_chart(grouped_data, "Medium", "Medium Activation Emotions")
plot_radar_chart(grouped_data, "Low", "Low Activation Emotions")
```

```python
import pandas as pd
import numpy as np

def compute_adorescore(file_path, output_path):
    # Load dataset
    df = pd.read_csv(file_path)

    # Define sentiment weights (adjustable for fine-tuning)
    emotion_weights = {
        "joy": 1.0, "trust": 0.8, "anticipation": 0.7, "surprise": 0.6,
        "sadness": -1.0, "fear": -0.8, "anger": -0.9, "disgust": -0.7,
        "ecstasy": 1.2, "admiration": 1.1, "terror": -1.2, "loathing": -1.1
    }

    # Process emotions and activation levels
    df['Detected_Emotions'] = df['Detected_Emotions'].str.split(', ')
    df['Confidence_Scores'] = df['Confidence_Scores'].str.split(', ').apply(lambda x: list(map(float, x)))

    # Compute emotion-weighted score for each row
    def calculate_score(emotions, confidences):
        score = 0
        for emotion, confidence in zip(emotions, confidences):
            weight = emotion_weights.get(emotion.lower(), 0)  # Default weight 0 if emotion not in dict
            score += weight * confidence
        return score * 100 / len(emotions) if emotions else 0  # Normalize to -100 to +100 range

    df['Adorescore'] = df.apply(lambda row: calculate_score(row['Detected_Emotions'], row['Confidence_Scores']), axis=1)

    # Save the result as CSV
    df[['Text', 'Adorescore', 'Main_Topic']].to_csv(output_path, index=False)
    print(f"Adorescore results saved to: {output_path}")

    return df[['Text', 'Adorescore', 'Main_Topic']]

# File paths
input_file = "/content/Final_Main_Subtopics.csv"
output_file = "Adorescore_Results.csv"

# Run Adorescore calculation and save results
result_df = compute_adorescore(input_file, output_file)
```

```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from transformers import pipeline
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
import re
import pickle

# Load models from .pkl files
with open('sia.pkl', 'rb') as f:
    sia = pickle.load(f)

with open('emotion_pipeline.pkl', 'rb') as f:
    emotion_pipeline = pickle.load(f)

with open('topic_classifier.pkl', 'rb') as f:
    topic_classifier = pickle.load(f)

with open('embedding_model.pkl', 'rb') as f:
    embedding_model = pickle.load(f)

with open('bertopic_model.pkl', 'rb') as f:
    bertopic_model = pickle.load(f)

#Load the processed text that was used to train the model.
df = pd.read_csv("Final_Main_Subtopics.csv")
processed_texts = df['Processed_Text'].astype(str).tolist()

#Refit the bertopic model.
bertopic_model.fit(processed_texts)

# Define topic lists
possible_topics = [
    "Product Quality", "Durability", "Defective Item", "Wrong Item Sent", "Packaging Issue",
    "Late Delivery", "Fast Delivery", "Damaged During Shipping",
    "Customer Support Response", "Refund & Returns", "Replacement Process",
    "Expensive Product", "Discounts & Deals", "Value for Money",
    "Website Usability", "Trust in Brand", "Fake or Counterfeit Product", "Delivery", "Quality", "Clothes"
]

def analyze_review(review_text):
    # Text preprocessing
    review_text = review_text.lower()
    review_text = re.sub(r'\W', ' ', review_text)
    review_text = re.sub(r'\s+', ' ', review_text).strip()

    # Sentiment analysis
    sentiment_score = sia.polarity_scores(review_text)['compound']

    # Emotion detection
    emotions_data = emotion_pipeline(review_text[:512])[0]
    primary_emotion = emotions_data[0]
    secondary_emotion = emotions_data[1]

    # Topic classification
    main_topics = topic_classifier(review_text, possible_topics, multi_label=True)
    selected_main_topics = [topic for topic, score in zip(main_topics['labels'], main_topics['scores']) if score > 0.5]
    if not selected_main_topics:
        selected_main_topics = ["General Feedback"]

    # Subtopic extraction
    topics, _ = bertopic_model.transform([review_text])
    sub_topic_labels = [
        ", ".join([word for word, _ in bertopic_model.get_topic(topic_id)])
        if topic_id != -1 else "Unknown"
        for topic_id in topics
    ]
    subtopics = {}
    for main_topic in selected_main_topics:
        subtopics[main_topic] = [sub_topic_labels[0]]

    # Adorescore calculation (simplified)
    emotion_weights = {
        "joy": 1.0, "trust": 0.8, "anticipation": 0.7, "surprise": 0.6,
        "sadness": -1.0, "fear": -0.8, "anger": -0.9, "disgust": -0.7,
        "ecstasy": 1.2, "admiration": 1.1, "terror": -1.2, "loathing": -1.1
    }
    adorescore_overall = 0
    for emotion in emotions_data:
        weight = emotion_weights.get(emotion['label'].lower(), 0)
        adorescore_overall += weight * emotion['score']
    adorescore_overall = round(adorescore_overall * 100)

    adorescore_breakdown = {}
    for main_topic in selected_main_topics:
        topic_words = main_topic.lower().split()
        topic_score = 0
        word_count = 0
        for word in topic_words:
            if word in review_text:
                topic_score += sentiment_score * 100
                word_count +=1
        if word_count>0:
            adorescore_breakdown[main_topic] = round(topic_score/word_count)
        else :
            adorescore_breakdown[main_topic] = round(sentiment_score*100)

    # Construct the result dictionary
    result = {
        "emotions": {
            "primary": {
                "emotion": primary_emotion['label'],
                "activation": "High" if primary_emotion['score'] > 0.75 else "Medium" if primary_emotion['score'] > 0.5 else "Low",
                "intensity": round(primary_emotion['score'], 2)
            },
            "secondary": {
                "emotion": secondary_emotion['label'],
                "activation": "High" if secondary_emotion['score'] > 0.75 else "Medium" if secondary_emotion['score'] > 0.5 else "Low",
                "intensity": round(secondary_emotion['score'], 2)
            }
        },
        "topics": {
            "main": selected_main_topics,
            "subtopics": subtopics
        },
        "adorescore": {
            "overall": adorescore_overall,
            "breakdown": adorescore_breakdown
        }
    }
    return result

# Example usage
review_text = "The delivery was fast, but the clothes didn't fit well and the quality was disappointing."
analysis_result = analyze_review(review_text)
print(analysis_result)
```
